```python
from pyspark.sql.functions import col, split, to_date, date_format

# Assuming you have a DataFrame named final_df

# Split 'CRASH DATE' column by 'T' and retrieve the 0 index
df = df.withColumn("CRASH DATE", split(col("CRASH DATE"), "T").getItem(0))

# Convert 'CRASH DATE' to date format
df = df.withColumn("CRASH DATE", to_date(col("CRASH DATE"), "yyyy-MM-dd"))

# Format 'CRASH DATE' as 'MM/DD/YYYY'
df = df.withColumn("CRASH DATE", date_format(col("CRASH DATE"), "MM/dd/yyyy"))
```

```python
df_csv = df_csv.select(
    col("CRASH DATE"),
    col("CRASH TIME"),
    col("BOROUGH"),
    col("ZIP CODE"),
    col("LATITUDE"),
    col("LONGITUDE"),
    col("NUMBER OF PERSONS INJURED"),
    col("NUMBER OF PERSONS KILLED"),
    col("NUMBER OF PEDESTRIANS INJURED"),
    col("NUMBER OF PEDESTRIANS KILLED"),
    col("NUMBER OF CYCLIST INJURED"),
    col("NUMBER OF CYCLIST KILLED"),
    col("NUMBER OF MOTORIST INJURED"),
    col("NUMBER OF MOTORIST KILLED"),
    col("CONTRIBUTING FACTOR VEHICLE 1"),
    col("CONTRIBUTING FACTOR VEHICLE 2"),
    col("CONTRIBUTING FACTOR VEHICLE 3"),
    col("CONTRIBUTING FACTOR VEHICLE 4"),
    col("CONTRIBUTING FACTOR VEHICLE 5"),
    col("COLLISION_ID"),
    col("VEHICLE TYPE CODE 1"),
    col("VEHICLE TYPE CODE 2"),
    col("VEHICLE TYPE CODE 3"),
    col("VEHICLE TYPE CODE 4"),
    col("VEHICLE TYPE CODE 5"))
```

```python
final_df=df.union(df_csv).dropDuplicates()
display(final_df)
```

```python
final_df.createOrReplaceTempView("final_df")
from pyspark.sql.functions import *
```

```python
result_df = spark.sql("""
    SELECT *, 
           ROW_NUMBER() OVER (PARTITION BY COLLISION_ID ORDER BY `CRASH TIME`) AS row_no
    FROM final_df
""")

# Filter the result DataFrame where row_no < 2
filtered_result_df = result_df.filter(result_df["row_no"] < 2)
```

```python
columns_to_keep = filtered_result_df.columns[:-1]

final_df = filtered_result_df.select(columns_to_keep)
```

```python
display(final_df)
```

```python
from pyspark.sql import functions as F

final_df = final_df.withColumn('BOROUGH', F.when(final_df['BOROUGH'].isNull(), 'Unknown').otherwise(final_df['BOROUGH']))

count_columns = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
                 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
                 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED']
for col in count_columns:
    final_df = final_df.withColumn(col, F.when(final_df[col].isNull(), '0').otherwise(final_df[col]))

final_df = final_df.withColumn('ZIP CODE', F.when(final_df['ZIP CODE'].isNull(), '10001').otherwise(final_df['ZIP CODE']))

final_df = final_df.withColumn('LATITUDE', F.when(final_df['LATITUDE'].isNull(), '40.4519').otherwise(final_df['LATITUDE']))
final_df = final_df.withColumn('LONGITUDE', F.when(final_df['LONGITUDE'].isNull(), '73.5826').otherwise(final_df['LONGITUDE']))
final_df = final_df.withColumn('VEHICLE TYPE CODE 1', F.when(final_df['VEHICLE TYPE CODE 1'].isNull(), 'UNKNOWN').otherwise(final_df['VEHICLE TYPE CODE 1']))
final_df = final_df.withColumn('CONTRIBUTING FACTOR VEHICLE 1', F.when(final_df['CONTRIBUTING FACTOR VEHICLE 1'].isNull(), 'Unspecified').otherwise(final_df['CONTRIBUTING FACTOR VEHICLE 1']))
final_df = final_df.withColumn('CRASH TIME', F.when(final_df['CRASH TIME'].isNull(), '00:00').otherwise(final_df['CRASH TIME']))
final_df = final_df.withColumn('COLLISION_ID', F.when(final_df['COLLISION_ID'].isNull(), '1001').otherwise(final_df['COLLISION_ID']))

# Round latitude and longitude to 2 decimal points
final_df = final_df.withColumn('LATITUDE', F.round(final_df['LATITUDE'], 2))
final_df = final_df.withColumn('LONGITUDE', F.round(final_df['LONGITUDE'], 2))
```

```python
# Save DataFrame to Parquet format
final_df.write.mode("overwrite").parquet("Files/Cleaned_Data")
```
